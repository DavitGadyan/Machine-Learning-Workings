# Visualizing Survived VS Not survived by Fare

```python
train.head()
```

```python
train.columns
```

```python
train.info()
```

```python
import seaborn as sns
train1=train.copy()
train1['Survived'].replace(1,'Survived',inplace=True)
train1['Survived'].replace(0,'Not Survived',inplace=True)
sns.lmplot( x='Age', y='Fare', data=train1, fit_reg=False, hue='Survived', legend=True,size=15)
plt.legend(loc='lower right')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Survival graph based on Age and Fare')

```

#### As we may clearly observe chnces of surviving were more if passenger were younger (more yellow dots on the left side where age is smaller than on right side)

#### Another interesting point is that passenger who paid higher fare in many cases has survived
